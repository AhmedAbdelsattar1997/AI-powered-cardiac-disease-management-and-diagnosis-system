```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder  # Encodes categorical data (e.g., strings) into numerical format.
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
# mean_squared_error < Measures the average squared difference between the actual and predicted values.
# r2_score Measures < how well the regression model fits the data. Values range from 0 (poor fit) to 1 (perfect fit).
from sklearn.impute import SimpleImputer

# Load datasets
df = pd.read_csv('heart_disease_uci.csv')

# Preview the expanded dataset
df.head()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cols = ['sex','cp','fbs','restecg','exang','slope','thal','dataset']

for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
x = df[["age","sex","cp","trestbps","chol","fbs","restecg","thalch","exang","oldpeak","slope","ca","thal"]]
y = df["num"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model = RandomForestClassifier(n_estimators=250,max_depth=10,random_state=42)

model.fit(x_train, y_train)

pred = model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, pred))
```

```python
import pandas as pd
import numpy as np
from flask import Flask, request, render_template_string
import traceback

app = Flask(__name__)

OPTIONS = {
    'sex': [('1', 'ذكر (Male)'), ('0', 'أنثى (Female)')],
    'cp': [('0', 'Typical Angina'), ('1', 'Atypical Angina'), ('2', 'Non-anginal Pain'), ('3', 'Asymptomatic')],
    'fbs': [('1', 'نعم (> 120 mg/dl)'), ('0', 'لا')],
    'restecg': [('0', 'Normal'), ('1', 'ST-T wave abnormality'), ('2', 'Left ventricular hypertrophy')],
    'exang': [('1', 'نعم'), ('0', 'لا')],
    'slope': [('0', 'Upsloping'), ('1', 'Flat'), ('2', 'Downsloping')],
    'thal': [('0', 'Normal'), ('1', 'Fixed Defect'), ('2', 'Reversible Defect')],
    'ca': [('0', '0'), ('1', '1'), ('2', '2'), ('3', '3')]
}

FEATURES_ORDER = ["age","sex","cp","trestbps","chol","fbs","restecg","thalch","exang","oldpeak","slope","ca","thal"]

html_code = """
<!DOCTYPE html>
<html dir="rtl">
<head>
    <title>Heart Disease AI</title>
    <style>
        body { font-family: 'Segoe UI', sans-serif; background: #eef2f3; padding: 20px; }
        .container { max-width: 600px; margin: auto; background: white; padding: 25px; border-radius: 15px; box-shadow: 0 10px 25px rgba(0,0,0,0.1); }
        h2 { text-align: center; color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 10px; }
        .form-grid { display: grid; grid-template-columns: 1 white; gap: 15px; margin-top: 20px; }
        label { font-weight: bold; color: #34495e; font-size: 14px; }
        input, select { width: 100%; padding: 10px; border: 1px solid #dcdde1; border-radius: 8px; box-sizing: border-box; }
        button { background: #3498db; color: white; border: none; padding: 15px; cursor: pointer; width: 100%; border-radius: 8px; font-size: 18px; transition: 0.3s; margin-top: 20px; }
        button:hover { background: #2980b9; }
        .res-box { margin-top: 25px; padding: 20px; border-radius: 10px; text-align: center; font-size: 18px; }
        .danger { background: #ff7675; color: white; }
        .safe { background: #55efc4; color: #2d3436; }
    </style>
</head>
<body>
    <div class="container">
        <h2>نظام توقع أمراض القلب الذكي</h2>
        <form method="POST" action="/predict">
            <div class="form-grid">
                {% for col in all_cols %}
                    <div>
                        <label>{{ col }}</label>
                        {% if col in options %}
                            <select name="{{ col }}">
                                {% for val, text in options[col] %}
                                    <option value="{{ val }}">{{ text }}</option>
                                {% endfor %}
                            </select>
                        {% else %}
                            <input type="number" step="any" name="{{ col }}" value="0">
                        {% endif %}
                    </div>
                {% endfor %}
            </div>
            <button type="submit">بدء التحليل باستخدام الذكاء الاصطناعي</button>
        </form>

        {% if prediction is not none %}
            {% if prediction == 0 %}
                <div class="res-box safe">النتيجة: لا يوجد مؤشرات لمرض القلب (سليم)</div>
            {% else %}
                <div class="res-box danger">النتيجة: هناك احتمالية إصابة (المستوى: {{ prediction }})</div>
            {% endif %}
        {% endif %}
    </div>
</body>
</html>
"""

@app.route('/')
def index():
    return render_template_string(html_code, all_cols=FEATURES_ORDER, options=OPTIONS, prediction=None)

@app.route('/predict', methods=['POST'])
def predict():
    try:
        data = [float(request.form.get(col, 0)) for col in FEATURES_ORDER]
        final_features = np.array([data])
        prediction = model.predict(final_features)[0]
        return render_template_string(html_code, all_cols=FEATURES_ORDER, options=OPTIONS, prediction=int(prediction))
    except Exception as e:
        print(traceback.format_exc())
        return f"حدث خطأ في البيانات: {e}"

if __name__ == '__main__':
    from google.colab import output
    output.serve_kernel_port_as_window(5000, path='/')
    app.run(port=5000)
```
